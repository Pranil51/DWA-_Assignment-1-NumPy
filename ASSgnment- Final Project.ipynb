```python
# 4. Filter the dataframe with only this columns
df=df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]
df
```

```python
#  a. Remove all duplicates observations

df.drop_duplicates(inplace=True)
df
```

```python
#  b. Find missing values in all columns
df.isnull()
```

```python
# c. Remove all observations where continent column value is missing
df.dropna(subset=['continent'],inplace=True)
df
```

```python
# d. Fill all missing values with 0
df.fillna(value=0,axis=1,inplace=True)
df
```

```python
# a. Convert date column in datetime format using pandas.to_datetime
df['date']=pd.to_datetime(df['date'])
df
```

```python
#b. Create new column month after extracting month data from date column.
df['month']=pd.DatetimeIndex(df['date']).month
df
```

```python
# 7. Data Aggregation:
#  a. Find max value in all columns using groupby function on 
# 'continent'
#  column
#  Tip: use reset_index() after applying groupby
#  b. Store the result in a new dataframe named 'df_groupby'.
#  (Use df_groupby dataframe for all further analysis)
```

```python
df_groupby=df.groupby(['continent']).total_cases.max().reset_index(name='max_total_cases')
df_groupby['max_total_deaths']=df.groupby(['continent']).total_deaths.max().reset_index(name='max_total_deaths')['max_total_deaths']
df_groupby['max_total_deaths']=df.groupby(['continent']).total_deaths.max().reset_index(name='max_total_deaths')['max_total_deaths']
df_groupby['max_gdp_per_capita']=df.groupby(['continent']).gdp_per_capita.max().reset_index(name='max_gdp_per_capita')['max_gdp_per_capita']
df_groupby['max_human_development_index']=df.groupby(['continent']).human_development_index.max().reset_index(name='max_human_development_index')['max_human_development_index']
df_groupby

```

```python
# 8. Feature Engineering :
#  a. Create a new feature 'total_deaths_to_total_cases' by ratio of
#  'total_deaths' column to 'total_cases'
```

```python
df_groupby['total_deaths_to_total_cases']=(df_groupby['max_total_deaths'])/(df_groupby['max_total_cases'])
```
